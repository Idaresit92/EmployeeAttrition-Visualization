```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/HR Employee Attrition.csv')
df
```

```python
df.info()
```

```python
#checking if there is null values in the dataset
df.isnull().sum()
```

```python
Num_Dept = df.groupby('Department').sum('EmployeeCount')
Num_Dept
```

```python
P = sns.barplot(data=Num_Dept.reset_index().sort_values('EmployeeCount', ascending=False), x='Department', y = 'EmployeeCount')
P.set(title = 'Total number of employees per department')
```

Above barplot shows Research & Development has the number of employees while Human resources has the least number of employees.

```python
Education_Field_count = df.groupby('EducationField').sum('EmployeeCount')
Education_Field_count
```

```python
P = sns.barplot(data=Education_Field_count.reset_index().sort_values('EmployeeCount', ascending=False), x='EducationField', y = 'EmployeeCount')
P.set(title = 'Total number of employees per EducationField')
```
